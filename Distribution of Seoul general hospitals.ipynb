## 1번 양재형

### value_counts

```python
cond = df['상권업종소분류명'] == '약국' # 데이터프레임에서 상권업종소분류명이 약국인 데이터만 추출하기 위한 조건식 설정
result = df[cond]['시도명'].value_counts()
print(result.to_frame())
```

### groupby

```python
cond = df['상권업종소분류명'] == '약국'
result = df[cond].groupby('시도명')['상권업종소분류명'].count().sort_values(ascending = False)
print(result)
```

### pivot_table

```python
cond = df['상권업종소분류명'] == '약국'
result = df[cond].pivot_table(index = '시도명', values = '상권업종소분류명',
                              aggfunc = 'count').sort_values(by = '상권업종소분류명', ascending = False)
print(result)
```

## 2번 권은서

```python
#지역별 동물병원 비율
df_vet_num = df[df["상권업종소분류명"]=="동물병원"]["시도명"].value_counts(normalize=True)

#지역별 약국 비율
df_phar_num = df[df["상권업종소분류명"]=="약국"]["시도명"].value_counts(normalize=True)

#지역별 병원 비율
df_hospital_num = df[df["상권업종중분류명"]=="병원"]["시도명"].value_counts(normalize=True)
```

```python
df_total = pd.DataFrame({
    '동물병원': df_vet_num,
    '약국': df_phar_num,
    '병원': df_hospital_num
})

# 그룹 막대그래프 생성
df_total.plot.bar(figsize=(14, 8), grid=True)

# 그래프 제목 및 축 레이블 설정
plt.title('동물병원, 약국, 병원의 지역별 비율', fontsize=16)
plt.xlabel('지역', fontsize=14)
plt.ylabel('비율', fontsize=14)

# 레전드(범례) 위치 설정
plt.legend(title='업종', title_fontsize=12, fontsize=12)

# x축 라벨 회전
plt.xticks(rotation=45, ha='right')

# 그래프 표시
plt.show()

#병원, 약국, 동물병원이 모두 서울특별시와 경기도에 집중된 것을 확인할 수 있음.
```

# 3번 신지예

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/corazzon/boostcourse-ds-510/master/data/medical_201909.csv.zip", low_memory=False)

df_s_h=df[(df['상권업종중분류명']=='병원') & (df['시도명']=='서울특별시')]
df_d = df_s_h[df_s_h["상권업종소분류명"].str.contains("피부") | df_s_h["상권업종소분류명"].str.contains("성형")]

df_r = df_d["시군구명"].value_counts() / df_s_h['시군구명'].value_counts()
df_r_s = df_r.sort_values(ascending=False)
df_r_s.round(2)
```

# 3번 박시연

```python
# 서울의 피부/성형 병원
seoul_pibu_hospital = df[(df["시도명"] == "서울특별시") & (df["상권업종중분류명"] == "병원") & (df['상권업종소분류명'].str.contains("피부|성형"))]

# 서울의 병원
seoul_hospital = df[(df["시도명"] == "서울특별시") & (df["상권업종중분류명"] == "병원")]
```

```python
# `피부과 성형외과 수 / 전체병원 수` 로 비율
ratio_seoul_pibu_hospital = seoul_pibu_hospital["시군구명"].value_counts() / seoul_hospital["시군구명"].value_counts()

# 소수점 두자리수 반올림, 내림차순 정렬
ratio_seoul_pibu_hospital.round(2).sort_values(ascending=False)
```

# 4번 - 김상은

```python
# 4번 김상은
import pandas as pd
import folium
from folium.plugins import MarkerCluster, HeatMap, MiniMap

# 데이터 로드
df = pd.read_csv("https://raw.githubusercontent.com/corazzon/boostcourse-ds-510/master/data/medical_201909.csv.zip", low_memory=False)

# 노인/치매병원 데이터 필터링
hospitals = df[(df['상권업종대분류명'] == '의료') & (df['상권업종소분류명'] == '노인/치매병원')]

# 대한민국 중심 좌표
center_lat, center_lon = 37.5642135, 127.0016985

# 지도 생성
m = folium.Map(location=[center_lat, center_lon], zoom_start=7, tiles='cartodbdark_matter')

# 마커 클러스터 생성
marker_cluster = MarkerCluster(name="노인/치매병원").add_to(m)

# 병원 위치 마커 추가
for idx, row in hospitals.iterrows():
    folium.CircleMarker(
        location=[row['위도'], row['경도']],
        radius=5,
        popup=row['상호명'],
        tooltip=row['상호명'],
        color='#FF9999',
        fill=True,
        fill_color='#FF9999'
    ).add_to(marker_cluster)

# 히트맵 레이어 추가
heat_data = [[row['위도'], row['경도']] for idx, row in hospitals.iterrows()]
HeatMap(heat_data, name="밀집도", radius=15).add_to(m)

# 미니맵 추가
MiniMap(toggle_display=True).add_to(m)

# 전체화면 버튼 추가
folium.plugins.Fullscreen(position='topright').add_to(m)

# 지도 타일 레이어 추가
folium.TileLayer('cartodbpositron', name="밝은 지도").add_to(m)
folium.TileLayer('cartodbdark_matter', name="어두운 지도").add_to(m)

# 레이어 컨트롤 추가
folium.LayerControl().add_to(m)

# 대한민국 행정구역 경계 추가
url = 'https://raw.githubusercontent.com/southkorea/southkorea-maps/master/kostat/2013/json/skorea_provinces_geo_simple.json'
folium.GeoJson(
    url,
    name='행정구역',
    style_function=lambda feature: {
        'fillColor': '#00000000',
        'color': 'yellow',
        'weight': 2,
        'fillOpacity': 0.1,
    }
).add_to(m)

# 지도 표시
m
```
